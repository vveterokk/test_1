```python
class Chocolate:
  def __init__(self,price,addition):
    self.price = (float(price))
    self.addition = (str(addition))
    if price > 0:
      print ('Цена больше 0')
    else:
      raise ValueError ('Цена должна быть больше 0!')
  def _show_my_chocolate_(self):
    if self.addition > ' ':
      print('Шоколадный батончик, добавка:', self.addition)
    else:
      print('Обычный шоколадный батончик')
  def __str__(self):
    return f'Шоколадный батончик, {self.price} рублей'
choco = Chocolate(199,'орех')
choco._show_my_chocolate_()
choco1 = Chocolate(5,' ')
choco1._show_my_chocolate_()
print(choco)


```
